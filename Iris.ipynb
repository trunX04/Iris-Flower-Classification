```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/iris.csv',header=None)
df = df.drop(0)
df
```

```python
df.columns=["Sepallengthincm","Sepalwidthincm","Petallengthincm","Petalwidthincm","Class"]
df
```

```python
X = df.drop(["Class"],axis=1)
y = df["Class"]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
import sklearn.metrics

labelsss = ['Versicolor', 'Setosa', 'Virginica']
print(sklearn.metrics.confusion_matrix(y_test, y_pred, labels=labelsss))
```
